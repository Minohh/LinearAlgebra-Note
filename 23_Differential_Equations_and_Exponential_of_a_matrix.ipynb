# Differential Equations
## Problem
Consider a vector $u$ with $n$ elements, and the vector will change as time gose on. So the vector has the form
$$u(t)=\begin{bmatrix}u_1(t)\\u_2(t)\\ \vdots\\u_n(t) \end{bmatrix}$$

Now suppose that we have vector at the time $t=0$
$$u(0)=\begin{bmatrix}1\\0\end{bmatrix}$$
And the differential of $u$ on $t$ is given by
$$\begin{align*}\frac{du_1}{dt}&=-u_1+2u_2\\
\frac{du_2}{dt}&=\ \ u_1-2u_2\end{align*}$$

## Solution 1
These two equations can be written in the form
$$u'=\begin{bmatrix}-1&2\\1&-2\end{bmatrix}\begin{bmatrix}u_1\\u_2\end{bmatrix}=Au$$
The eigenvalues and eigenvectors of matrix $A$ is
$$\lambda_1=0,\lambda_2=-3\qquad x_1=\begin{bmatrix}2\\1\end{bmatrix},x_2=\begin{bmatrix}1\\-1\end{bmatrix}$$
There is two different eigenvalues, so we the <font color='red'>general solution</font> of the differential equation is given by
<font color='red'>$$u(t)=c_1e^{\lambda_1t}x_1+c_2e^{\lambda_2t}x_2$$</font>
Substitude the eigenvalues and eigenvectors to this general solution
$$u(t)=c_1 e^0 \begin{bmatrix}2\\1\end{bmatrix}+c_2 e^{-3t} \begin{bmatrix}1\\-1\end{bmatrix}$$
We use the initial conditions to get $c_1,c_2$
$$u(0)=c_1 \begin{bmatrix}2\\1\end{bmatrix}+c_2 \begin{bmatrix}1\\-1\end{bmatrix}=\begin{bmatrix}1\\0\end{bmatrix}$$
The solution of $c_1,c_2$ is 
$$c_1=c_2=\frac{1}{3}$$
Thus 
$$u(t)=\frac{1}{3} \begin{bmatrix}2\\1\end{bmatrix}+\frac{1}{3} e^{-3t} \begin{bmatrix}1\\-1\end{bmatrix}$$
For the limit case
$$u(\infty)=\frac{1}{3} \begin{bmatrix}2\\1\end{bmatrix}+\frac{1}{3} e^{-3t} \begin{bmatrix}1\\-1\end{bmatrix}=\frac{1}{3} \begin{bmatrix}2\\1\end{bmatrix}$$
If $t$ tends to $\infty$, the second term tends to zero because the negative value $\lambda_2$ makes $e^{\lambda_2t}$ to go to zero. 

## <font color='red'>Solution 2</font>
This solution can be seen as an explanation of the solution above.  
$$\frac{du}{dt}=Au$$
Diagonalizing $A$, we receive the eigenvalues and eigenvectors. combine them to build the matrix $S$, $S^{-1}$ and $\Lambda$.  
Now let $v=S^{-1}u$, or we say $u=Sv$, then the differential equation become
$$\frac{du}{dt}=S\frac{dv}{dt}=ASv$$
Multiply both side by $S^{-1}$
$$\frac{dv}{dt}=S^{-1}ASv=\Lambda v$$
which can be written to
$$\begin{bmatrix}\frac{dv_1}{dt}\\ \frac{dv_2}{dt}\\ \vdots\\ \frac{dv_n}{dt}\end{bmatrix}=
\begin{bmatrix}\lambda_1 &0 &\cdots &0\\ 0 &\lambda_2 &\cdots &0\\ \vdots &\vdots &\ddots &\vdots\\ 0 &0 &\cdots &\lambda_n\end{bmatrix} 
\begin{bmatrix}v_1\\ v_2\\ \vdots\\ v_n\end{bmatrix}
=\begin{bmatrix}\lambda_1 v_1\\ \lambda_2v_2\\ \vdots\\ \lambda_nv_n\end{bmatrix}$$
Each element is decoupled, thus we can use the <font color='red'>general differential solution</font> to solve the equation of each row.
<font color='red'>$$v_n=c_ne^{\lambda_n t}$$</font>
Finally
$$u=Sv$$

# Convergence
## Stability
How $\displaystyle{\lim_{t\to\infty}u(t)=0}$
- u(t) has a general form that is $u(t)=c_1e^{\lambda_1t}x_1+c_2e^{\lambda_2t}x_2+\cdots$, if we want $u(t)$ to be zero as $t$ goes to $\infty$, then all the eigenvectors should be negative $\lambda<0$.
- For the case that the eigenvalue is a complex value $\lambda=a+bi$, then $e^{\lambda t}=e^{(a+bi)t}=e^{at}e^{bti}$. The absolute value is $|e^{\lambda t}|=|e^{at}||e^{bti}|=e^{at}$, where the absolute value of imaginal part is 1, and $bt$ denotes the angle from the vector to the real coordinate. It's the real part that matters. Real part of $\lambda$ has to be negative.

## Steady state
How $\displaystyle{\lim_{t\to\infty}u(t)=c}$
$$\lambda_1=0, other\ \lambda<0$$

## Blow up (Divergence)
How $\displaystyle{\lim_{t\to\infty}u(t)}$ divergence.  
$$If\ any\ \lambda>0$$

# Exponentials
Tylor Series
$$e^{At}=I+At+\frac{(At)^2}{2}+\frac{(At)^3}{6}+\cdots+\frac{(At)^n}{n!}+\cdots$$
If $A$ can be diagonalized, then we get
$$\begin{align*}e^{At}&=SIS^{-1}+S\Lambda S^{-1}t+\frac{S\Lambda^2 S^{-1}t^2}{2}+\cdots\\
&=S\left(I+\Lambda t+\frac{\Lambda^2t^2}{2}+\cdots\right)S^{-1}\\
&=Se^{\Lambda t}S^{-1}\end{align*}$$
where $\Lambda$ is the diagonal matrix, then the matrix $e^{\Lambda t}$ is
$$e^{\Lambda t}=I+\Lambda t+\frac{\Lambda^2t^2}{2}+\cdots
=\begin{bmatrix}
1+\lambda_1+\frac{\lambda_1^2t^2}{2}+\cdots &0 &\cdots &0\\
0 &1+\lambda_2+\frac{\lambda_2^2t^2}{2}+\cdots &\cdots &0\\
\vdots &\vdots &\ddots &\vdots\\
0 &0 &\cdots &1+\lambda_n+\frac{\lambda_n^2t^2}{2}+\cdots\\
\end{bmatrix}
=
\begin{bmatrix}
e^{\lambda_1 t} &0 &\cdots &0\\
0 &e^{\lambda_2 t} &\cdots &0\\
\vdots &\vdots &\ddots &\vdots\\
0 &0 &\cdots &e^{\lambda_n t}\\
\end{bmatrix}$$
If we want $e^{At}$ to go to zero while $t\to\infty$, all the $e^{\lambda t}$ should be zero, which means that all the $\lambda$ should be  negative.

# Approach to solving differential equation
Suppose there is a differential equation
$$y''+by'+ky=0$$
We want to change it to the form $u'=Au$, where the unknow vectors are
$$u=\begin{bmatrix}y'\\y\end{bmatrix},u'=\begin{bmatrix}y''\\y'\end{bmatrix}$$
We change the form above to $y''=-by'-ky$. For solving the 2-variable formular we need another more equation which is $y'=y'+0\cdot y$. 
$$\left\{\begin{align*}
y''&=-by'-ky \\ 
y'&=y' 
\end{align*}\right.$$
Change the form to matrix multiplication
$$u'=Au=\begin{bmatrix}-b&-k\\1&0\end{bmatrix}\begin{bmatrix}y'\\y\end{bmatrix}$$
Then we can diagonalize the matrix $A$ and solve the differential equation.

For the equations with multiple differentials
$$y^{(k)}+c_1y^{(k-1)}+\cdots+c_{k-1}y^{(1)}+c_ky=0$$
The matrix $A$ is
$$A=\begin{bmatrix}
c_1 &c_2 &\cdots &c_{k-1} &c_k\\
1 &0 &\cdots &0 &0\\
0 &1 &\cdots &0 &0\\
\vdots &\vdots &\ddots &\vdots &\vdots\\
0 &0 &\cdots &1 &0
\end{bmatrix}$$
