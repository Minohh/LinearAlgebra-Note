# Definition
Suppose that we have a transform from $n$-dim real space to $n$-dim real space, which is given by
$$T:\mathbb{R}^n\to\mathbb{R}^n$$
We input the vector $\vec{x}$, which will be transformed by multipling this vector by a  matrix $A$.
$$T(\vec{x})=A\vec{x}$$
If there is a vector $\vec{v}$ ($\vec{v}\neq \vec{0}$) that satisfy the form below
$$T(\vec{v})=A\vec{v}=\lambda \vec{v}$$
where $\lambda$ is a constant value, then we call the vector $\vec{v}$ *eigenvector*, $\lambda$ is the corresponding *eigenvalue*.

-------------------------
# Relationship

$$\begin{align*}\vec{0}&=\lambda\vec{v}-A\vec{v}\\
&=\lambda I_n\vec{v}-A\vec{v}\\
&=(\lambda I_n-A) \vec{v}
\end{align*}$$
gives
$$\vec{v}\in N(\lambda I_n-A) \Leftrightarrow det(\lambda I_n-A)=|\lambda I_n-A|=0$$
This means $\vec{v}$ is in the null space of $\lambda I_n-A$, which is equivalent to that the determinant of $\lambda I_n-A$ is $0$. Such result shows that each column of $\lambda I_n-A$ is linear dependent.  
linear dependent:
$$For\ a\ set\ of\ vectors\ \{\vec{v_1},\vec{v_2},\cdots,\vec{v_n}\},\quad \exists \{a_1,a_2,\cdots,a_n\},\quad a_1\vec{v_1}+a_2\vec{v_2}+\cdots+a_n\vec{v_n}=\vec{0}$$

**Conclusion**:
$A\vec{v}=\lambda\vec{v}$ for non-zero $\vec{v}$,if and ouly if $det(\lambda I_n-A)=0$.

-----------------
# Find eigenvalues and eigenvectors
In an other word, $\lambda$ is an eigenvalue of $A$ if and only if $det(\lambda I_n-A)=0$.
## $2\times 2$ matrix
### solving for the eigenvalues
Suppose we have a matrix
$$A=\begin{bmatrix}
1 &2 \\ 
4 &3 
\end{bmatrix}$$
$$\begin{align*}det(\lambda I_n-A)
&=det
\left(\begin{bmatrix}\lambda &0 \\ 0 & \lambda\end{bmatrix}
- \begin{bmatrix}1 &2 \\ 4 & 3\end{bmatrix}\right )\\
&=det\left( \begin{bmatrix}\lambda-1 &-2 \\ -4 & \lambda-3\end{bmatrix}\right )\\
&=(\lambda-1)(\lambda-3)-(-2)\times(-4)\\
&=\lambda^2-4\lambda-5\\
&=0
\end{align*}$$
solving this equation gives
$$\lambda = -1,5$$

### solving for the eigenvectors
Now we have our eigenvalues, the next step is to find the corresponding eigenvectors.
From the definition, we receive
$$(\lambda I_n-A)\vec{v}=\vec{0}$$
*Eigenspace* is the space of all the corresonding eigenvectors. Let $E_5$ be the *eigenspace* that correspond to the eigenvalue $\lambda=5$. $E_5$ is also equvalent to the null space of $5 I_n-A$, which is denotes by $N(5I_n-A)$.
$$E_5=N\left(\begin{bmatrix}5 &0 \\ 0 & 5\end{bmatrix}
-\begin{bmatrix}1 &2 \\ 4 & 3\end{bmatrix}\right)
=N\left(\begin{bmatrix}4 &-2 \\ -4 & 2\end{bmatrix}\right)$$
solving the equation $\begin{bmatrix}4 &-2 \\ -4 & 2\end{bmatrix}\begin{bmatrix}v_1\\v_2\end{bmatrix}=\begin{bmatrix}0\\0\end{bmatrix}$. We receive
$$E_5=\left\{\begin{bmatrix}v_1\\v_2\end{bmatrix}=t\begin{bmatrix}1/2\\1\end{bmatrix},\ t\in\mathbb{R}\right\}$$
Similarly,
$$E_{-1}=\left\{\begin{bmatrix}v_1\\v_2\end{bmatrix}=t\begin{bmatrix}-1\\1\end{bmatrix},\ t\in\mathbb{R}\right\}$$

## $3\times 3$ matrix
### solving for the eigenvalues
Suppose we have a matrix 
$$A=\begin{bmatrix}
-1 &2  &2 \\ 
2 & 2 &-1 \\ 
2 &-1  &2 
\end{bmatrix}$$
Follow the solving process above, we finally figure out that the eigenvalues are given by
$$\lambda=3,-3$$

### solving for the eigenvectors
$$\begin{align*}E_3&=\left\{\begin{bmatrix}v_1\\v_2\\v_3\end{bmatrix}=a\begin{bmatrix}1/2\\1\\0\end{bmatrix}+b\begin{bmatrix}1/2\\0\\1\end{bmatrix},\ a,b\in\mathbb{R}\right\}\\
E_{-3}&=\left\{\begin{bmatrix}v_1\\v_2\\v_3\end{bmatrix}=t\begin{bmatrix}-2\\1\\1\end{bmatrix},\ t\in\mathbb{R}\right\}\end{align*}$$

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.patches as mpatches

Orig = [0, 0]
V = np.array([[0.5, 1], [-1, 1]])
clr = ['C0', 'C1']

def axesCross(ax):
    ax.spines["top"].set_color("none")
    ax.spines["bottom"].set_position("zero")
    ax.spines["left"].set_position("zero")
    ax.spines["right"].set_color("none")

def set_limit(ax, limit):
    ax.set_xlim(-limit, limit)
    ax.set_ylim(-limit, limit)

def crop(x, limit):
    x[x>limit]  = np.nan
    x[x<-limit] = np.nan
    
def main():
    fig = plt.figure(figsize=(11,5), dpi=60)
    ax1 = fig.add_subplot(1,2,1)
    ax2 = fig.add_subplot(1, 2, 2, projection='3d')
    
    axesCross(ax1)
    limit = 3
    set_limit(ax1, 3)
    set_limit(ax2, 3)
    ax2.set_zlim(-limit, limit)
    
    x = np.linspace(-3,3,10)
    E5, = ax1.plot(x, 2*x, color='C0', label=r"$E_5$")
    Em1, = ax1.plot(x, -x, color='C1', label=r"$E_{-1}$")
              
    ax1.quiver(Orig[0], Orig[1], V[:, 0], V[:, 1], color=clr, angles='xy', scale=1, scale_units='xy')
    ax1.legend(handles=[E5, Em1])
    ax1.quiver(Orig[0], Orig[1], V[:, 0], V[:, 1], color=clr, angles='xy', scale=1, scale_units='xy')
    ax1.legend(handles=[E5, Em1])
    
    E31, = ax2.plot(0.5*x, x, np.zeros(len(x)), color='C0')
    E32, = ax2.plot(0.5*x, np.zeros(len(x)), x, color='C1')
    Em3, = ax2.plot(x, -0.5*x, -0.5*x, color='C2', label=r"$E_{-3}$")
    
    X,Y = np.mgrid[-3:3:700j,-3:3:700j]
    Z = 2.0*X - 1.0*Y
    Z[Z>3] = np.nan
    Z[Z<-3] = np.nan
    
    ax2.plot_surface(X,Y,Z, color='C3', alpha=0.5)
    
    #ax2.quiver(Orig[0], Orig[1], V[:, 0], V[:, 1], color=clr, angles='xy', scale=1, scale_units='xy')
    E3 = mpatches.Patch(color='C3', alpha=0.5, label=r"$E_3$")
    ax2.legend(handles=[E3, Em3])
    ax2.dist = 8


if __name__=="__main__":
    main()
```

# Usage
Consider a tranformation $T:\mathbb{R}\to\mathbb{R}$, which is given by
$$T(\vec{x}) = \underset{n\times n}{A}\vec{x}$$
Assume $A$ has n Linear Independent eigenvectors, and we put them together as columns to combine matrix $C$
$$C=[\vec{v_1},\vec{v_2},\cdots,\vec{v_n}]$$
Following the definition, we have n eigenvalues. 
$$\begin{matrix}
T(\vec{v_1}) &=A\vec{v_1}  &=\lambda_1\vec{v_1} \\ 
T(\vec{v_2}) &=A\vec{v_2}  &=\lambda_2\vec{v_2} \\ 
&\vdots\\
T(\vec{v_n}) &=A\vec{v_n}  &=\lambda_n\vec{v_n} 
\end{matrix}$$
which is 
$$T(C) = AC = [\lambda_1\vec{v_1},\lambda_2\vec{v_2},\cdots,\lambda_n\vec{v_n}]=[\vec{v_1},\vec{v_2},\cdots,\vec{v_n}]\begin{bmatrix}
\lambda_1 &0 &\cdots &0\\
0 &\lambda_2 &\cdots &0\\
&\vdots \\
0 &0 &\cdots &\lambda_n
\end{bmatrix}=CD$$
From this equation, we can receive that
$$A=CDC^{-1}$$
$$D=C^{-1}DC$$

Now we will introduce more details about this equation. Consider that each eigenvector is the base of coordinates of a new system.
For a vector $\vec{x}$, we perform the following steps
1. Project $\vec{x}$ to each base that is represented by $n$ eigenvectors. We use matrix $\underset{n\times n}{C^{-1}}$ to denote the projection whereas the inverse computation is denoted by $C$. And the projected component are denoted by matrix $\underset{n\times n}{[\vec{x}]_B}$
$$[\vec{x}]_B=C^{-1}\vec{x}$$
2. Transformation of $\vec{x}$ is equivalent to multiply each component by the corresponding eigenvalues
$$[T(\vec{x})]_B=D[\vec{x}]_B$$
where
$$D=\begin{bmatrix}
\lambda_1 &0 &\cdots &0\\
0 &\lambda_2 &\cdots &0\\
&\vdots \\
0 &0 &\cdots &\lambda_n
\end{bmatrix}$$
3. Combine all the components on bases
$$T(\vec{x})=C[T(\vec{x})]_B$$

Thus, it can be written
$$T(\vec{x})=CDC^{-1}\vec{x}$$
